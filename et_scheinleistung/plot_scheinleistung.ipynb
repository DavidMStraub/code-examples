# Demonstration Wirkleistung und Blindleistung

In diesem Notebook wird die Scheinleistung in einem Wechselstromkreis mit verschiedenen Phasenverschiebungen zwischen Spannung und Strom dargestellt. Mithilfe eines Sliders kann die Phasenverschiebung interaktiv angepasst werden, um die Auswirkungen auf die Wirkleistung und Blindleistung zu beobachten.

Um das Notebook auszuführen, werden die Bibliotheken `ipywidgets` und `ipympl` benötigt (`pip install ipywidgets ipympl`).


```python
%matplotlib widget

import scheinleistung
from math import pi
import matplotlib.pyplot as plt
from ipywidgets import FloatSlider, interactive_output
```

```python
fig, axes = plt.subplots(3, 1, figsize=(8, 8))

def update(phi_over_pi):
    scheinleistung.plot_scheinleistung(phi_over_pi * pi, ax=axes)
    fig.canvas.draw_idle()

slider = FloatSlider(
    min=-0.5,
    max=0.5,
    step=0.05,
    value=0.0,
    description='φ/π'
)

interactive_output(update, {'phi_over_pi': slider})
slider
```
